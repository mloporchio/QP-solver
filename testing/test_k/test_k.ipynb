```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('test_k_alg.csv', header=None);
```

```python
df1.columns = ['n', 'k', 'opt', 'it', 'time', 'ctol', 'dtol']
df1.drop(labels=['ctol', 'dtol'], axis=1, inplace=True)
df1.sort_values(ascending=True, by='k', inplace=True)
df1.head()
```

```python
plt.figure(figsize=(10, 7))
plt.title("Number of iterations (n = 1000, ecc = 0.5)")
plt.plot(df1['k'], df1['it'], color='black')
plt.plot([x for x in range(0, 1000, 10)], [1000-x for x in range(0, 1000, 10)], color='black', linestyle='--')
plt.ylabel("Iterations")
plt.xlabel("k")
plt.savefig("test_k.png", dpi=170, format='png', bbox_inches='tight')
```
