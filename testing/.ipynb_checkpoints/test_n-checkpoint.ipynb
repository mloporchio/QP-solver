```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('test_n_matlab.csv', header=None);
df2 = pd.read_csv('test_n_alg.csv', header=None);
```

```python
df1.columns = ['n', 'k', 'mat_opt', 'mat_it', 'mat_time']
df1.head()
```

```python
df2.columns = ['n', 'k', 'opt', 'it', 'time', 'ctol', 'dtol']
df2.drop(labels=['ctol', 'dtol'], axis=1, inplace=True)
df2.head()
```

```python
res = pd.merge(df1, df2, left_on=['n', 'k'], right_on=['n', 'k'])
res.sort_values(ascending=True, by='n', inplace=True)
res
```

```python
res.to_csv('test_n_full.csv', header=True, index=False)
```
