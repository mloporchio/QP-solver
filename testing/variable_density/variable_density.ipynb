```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('variable_density_matlab.csv', header=None);
df2 = pd.read_csv('variable_density_alg.csv', header=None);
```

```python
df1.columns = ['n', 'k', 'den', 'mat_opt', 'mat_it', 'mat_time']
df1.head()
```

```python
df2.columns = ['n', 'k', 'den', 'opt', 'it', 'time', 'ctol', 'dtol']
df2.drop(labels=['ctol', 'dtol'], axis=1, inplace=True)
df2.head()
```

```python
res = pd.merge(df1, df2, left_on=['n', 'k', 'den'], right_on=['n', 'k', 'den'])
res.sort_values(ascending=True, by='den', inplace=True)
res
```

```python
plt.figure(figsize=(10, 7))
plt.title("Running time (n = 1000, k = 500)")
plt.plot(res['den'], res['mat_time'], color='gray', label='Matlab')
plt.plot(res['den'], res['time'], color='black', label='PGM')
plt.ylabel("Time (ms)")
plt.xlabel("Density")
#plt.plot([x for x in range(100, 2000)], [x*2 for x in range(100, 2000)], color='black', linestyle='--')
plt.legend()
plt.savefig("test_density.png", dpi=170, format='png', bbox_inches='tight')
```

```python
res.to_csv('variable_density_full.csv', header=True, index=False)
```
